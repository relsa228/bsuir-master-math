### **Задача 1.** 
Минимизировать
$$
1 \space | \space | L_{max}
$$

Алгоритм: EDD (2.3)

```python
jobs = [
    {'id': 'Job1', 'p': 8, 'd': 12},
    {'id': 'Job2', 'p': 3, 'd': 7},
    {'id': 'Job3', 'p': 12, 'd': 23},
    {'id': 'Job4', 'p': 1, 'd': 4},
    {'id': 'Job5', 'p': 7, 'd': 21},
    {'id': 'Job6', 'p': 5, 'd': 17},
    {'id': 'Job7', 'p': 3, 'd': 8}
]
```

```python
sorted_schedule = sorted(jobs, key=lambda x: x['d'])

current_time = 0
l_max = -float('inf')
results = []
```

```python
for job in sorted_schedule:
    p_i = job['p']
    d_i = job['d']
    name = job['id']
    
    start = current_time
    current_time += p_i
    c_i = current_time
    l_i = c_i - d_i
    
    if l_i > l_max:
        l_max = l_i
        
    results.append({
        'ID': name,
        'Start': start,
        'C_i': c_i,
        'd_i': d_i,
        'L_i': l_i
    })
```

```python
print(f"{'ID':<6} | {'Начало':<7} | {'Конец':<7} | {'Дедлайн':<8} | {'L_i'}")
print("-" * 45)
for r in results:
    print(f"{r['ID']:<6} | {r['Start']:<7} | {r['C_i']:<7} | {r['d_i']:<8} | {r['L_i']}")

print("-" * 45)
print(f"Минимально возможное L_max = {l_max}")
```

### **Задача 2.** 
Минимизировать
$$
1 \space | \space | n_{t}
$$
Алгоритм: Мура-Ходжсона. (2.8)

```python
import heapq

sorted_jobs = sorted(jobs, key=lambda x: x['d'])

current_set = []
current_time = 0
rejected_jobs = []
```

```python
for job in sorted_jobs:
    p_i = job['p']
    d_i = job['d']
    name = job['id']
    
    current_time += p_i
    heapq.heappush(current_set, (-p_i, name))
    
    if current_time > d_i:
        longest_p, longest_id = heapq.heappop(current_set)
        longest_p = -longest_p

        current_time -= longest_p
        rejected_jobs.append(longest_id)
        print(f"Выбросили {longest_id} (p={longest_p}), так как из-за неё текущее время {current_time + longest_p} > дедлайна {d_i}")
```

```python
on_time_jobs = [item[1] for item in current_set]
```

```python
print("-" * 30)
print(f"Работы вовремя: {on_time_jobs}")
print(f"Опоздавшие работы: {rejected_jobs}")
print(f"Минимальное число опоздавших: {len(rejected_jobs)}")
```
