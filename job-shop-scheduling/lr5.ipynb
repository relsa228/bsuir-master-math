# Лабораторная работа 5. Вариант 1
**Тема.** Задачи с последовательным выполнением работы на нескольких станках

## **Задача 1**
$$
F_2 | \space | C_{max}
$$ 
Это новый тип задач ($F$), они предполагают, что работа должна пройти через несколько станков последовательно.

### Алгоритм
Правило Джонсона [1, 4.2] применяется для решения задач с последовательным выполнением на двух станках. Формулировка эвристики:
> Самые короткие операции — в начало или в конец

```python
jobs = [
    {'id': 'J1', 'M1': 1, 'M2': 13},
    {'id': 'J2', 'M1': 10, 'M2': 12},
    {'id': 'J3', 'M1': 17, 'M2': 9},
    {'id': 'J4', 'M1': 12, 'M2': 17},
    {'id': 'J5', 'M1': 11, 'M2': 3}
]
```

```python
left = []
right = []
remaining_jobs = list(jobs)
```

Цикл алгорима Джонсона состоит из нескольких шагов:
- шаг 1: найти минимальное время обработки среди всех оставшихся работ на обоих станках;
- шаг 2: если этот минимум относится к первому станку, поставьте эту работу в начало очереди, выход, иначе переход к **шагу 3**;
- шаг 3: если этот минимум относится ко второму станку, поставьте эту работу в конец очереди.

```python
while remaining_jobs:
    min_val = float('inf')
    best_job = None
    machine = None
    for job in remaining_jobs:
        if job['M1'] < min_val:
            min_val = job['M1']
            best_job = job
            machine = 1
        if job['M2'] < min_val:
            min_val = job['M2']
            best_job = job
            machine = 2
    if machine == 1:
        left.append(best_job['id'])
    else:
        right.insert(0, best_job['id'])
    remaining_jobs.remove(best_job)
```

Объединим очереди (начало и конец) и составим расписание

```python
optimal_order = left + right
print(f"Оптимальный порядок по Джонсону: {optimal_order}")
```
