**Задача 1.**
$$
J_2 | \space | C_{max}
$$ 

Алгоритм: Jackson's Rule

```python
j12 = [{'id': 'A', 'p1': 20, 'p2': 7}, {'id': 'B', 'p1': 8, 'p2': 12}]
j21 = [{'id': 'C', 'p1': 12, 'p2': 14}, {'id': 'D', 'p1': 22, 'p2': 10}, {'id': 'E', 'p1': 14, 'p2': 20}, {'id': 'F', 'p1': 16, 'p2': 11}]
j1 = [{'id': 'OnlyM1_1', 'p1': 17}, {'id': 'OnlyM1_2', 'p1': 9}]
j2 = [{'id': 'OnlyM2_1', 'p2': 16}, {'id': 'OnlyM2_2', 'p1': 13}]
```

```python
def solve_johnson_rule(jobs):
    left = []
    right = []
    remaining_jobs = list(jobs)
    
    while remaining_jobs:
        min_val = float('inf')
        best_job = None
        machine = None
        
        for job in remaining_jobs:
            if job['p1'] < min_val:
                min_val = job['p1']
                best_job = job
                machine = 1
            if job['p2'] < min_val:
                min_val = job['p2']
                best_job = job
                machine = 2
        
        if machine == 1:
            left.append(best_job['id'])
        else:
            right.insert(0, best_job['id'])
            
        remaining_jobs.remove(best_job)
        
    optimal_order = left + right
    return optimal_order
    
def solve_j2_jackson(jobs_12, jobs_21, jobs_1, jobs_2):
    s12 = solve_johnson_rule(jobs_12) 
    
    for j in jobs_21:
        j['p1'], j['p2'] = j['p2'], j['p1']
    s21 = solve_johnson_rule(jobs_21)

    m1_order = s12 + [j['id'] for j in jobs_1] + s21
    m2_order = s21 + [j['id'] for j in jobs_2] + s12
    
    return m1_order, m2_order

m1, m2 = solve_j2_jackson(j12, j21, j1, j2)
print(f"Очередь на M1: {m1}")
print(f"Очередь на M2: {m2}")
```
