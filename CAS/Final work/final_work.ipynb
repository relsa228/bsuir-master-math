**Задание 1.** Найдите 20 членов последовательности 
$
\{\frac{1}{n \cdot (n + 2)}\}^{\infty}_{n=1}
$

Объявим общий член ряда

```python
n = var('n')
a_n = 1 / (n * (n + 2))
show(a_n)
```

Найдем первые $ n $ членов ряда через подстановку константных чисел от 1 до 20. Подстановка осуществляется через `subs()`

```python
for i in range(1, 21):
    show(a_n.subs(n=i))
```

**Задание 2.** Найдите сумму ряда $ \sum^{\infty}_{n=1} \frac{1}{n^{102/100}} $ и ее числовое значение. Оцените отклонение полученного значения от частичных сумм $ S_{100} $ это ряда.

Объявим общий член ряда

```python
n = var('n')
a_n = 1 / (n ** (102/100))
show(a_n)
```

```python
exact_sum = sum(a_n, n, 1, oo, algorithm='sympy')
show(exact_sum)
```

Как видно, в результате была получена $ \zeta $-функция. Получим численное решения с ее помощью.

```python
zeta_sum = zeta(51/50).n()
show(zeta_sum)
```

Сравним полученный результат с частичной суммой $ S_{100} $

```python
n_min = 1
n_max = 100

s_100 = sum([a_n.subs(n=k).n() for k in range(n_min, n_max + 1)])
show(s_100)
```

Как видно частичная сумма серьезно отличается от фактической. Построим график частичных сумм и полной суммы.

```python
p_part_sums = list_plot([(k, a_n(n=k)) for k in range(n_min, n_max + 1)], color='blue', marker='o')
p_full_sum = list_plot([zeta_sum], color='red', marker='o')
show(p_part_sums + p_full_sum)
```
